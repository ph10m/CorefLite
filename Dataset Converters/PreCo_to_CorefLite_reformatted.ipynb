```python
dev = True

_original = "train.jsonl"
_new = "preco.coreflite"
if dev:
    _original = "dev.jsonl"
    _new = "preco_dev.coreflite"

in_file = os.path.join(datadir, _original)

outdir = "../../coreference_data"
out_file = os.path.join(outdir, _new)
```

```python
import jsonlines
```

```python
with jsonlines.open(out_file) as f:
    for obj in f:
        toks = obj["tokens"]
        clust = obj["clusters"]
        print(clust)
        for cl in clust:
            print("____")
            
            for m in cl:
                print(toks[m[0]:m[1]+1])
            print()
        print(toks)
        break
```

```python
with jsonlines.open(in_file) as f:
    for obj in f:
        sents = obj["sentences"]
        clusts = obj["mention_clusters"]
        for cl in clusts:
            print("____")
            for m in cl:
                sentid, start, end = m
                print(sents[sentid][start:end])
            print()
        print(sents)
        break
```
